```python
# This will import all the widgets
# and modules which are available in
# tkinter and ttk module
from tkinter import * 
from tkinter.ttk import *

#--------------------------------------------------------------------------------------------------------------------------------------------
# This function runs the scanner
def info():
    a = %run BookFinder2.ipynb
    return a

def quit():
    master.quit()

def runBar():
    %run Barcode_Reader_Final_Version.ipynb
    newWindow = Toplevel(master)
    newWindow.title("Result")
    newWindow.geometry("600x300")
    Label(newWindow, text = str(info())).pack()
    print("this")
    print (str(info()))
    print (type(info()))

    btn = Button(newWindow, 
             text ="Correct Book") #now add to database
    btn.pack(pady = 10)

    btn2 = Button(newWindow, 
             text ="Incorrect Book") #colapse all
    btn2.pack(pady = 10)
    

    #Run bookfinder and then open window with summary
#--------------------------------------------------------------------------------------------------------------------------------------------
'''MAIN WINDOW'''

# creates a Tk() object
master = Tk()

master.title("MSP Database")

# sets the geometry of main 
# root window
master.geometry("600x300")

#--------------------------------------------------------------------------------------------------------------------------------------------
'''ADD WINDOW'''

# function to open a new window 
# on a button click
def openAddBook():
      
    # Toplevel object which will 
    # be treated as a new window
    newWindow = Toplevel(master)
  
    # sets the title of the
    # Toplevel widget
    newWindow.title("Add a book to the MSP database")
  
    # sets the geometry of toplevel
    newWindow.geometry("600x300")
  
    # A Label widget to show in toplevel
    Label(newWindow, 
          text ="Insert Name then Scan Book").pack()

    name = Entry(newWindow,width=20) # to get this use name.get()
    name.pack(pady=10)

    btn = Button(newWindow, 
             text ="Scan Book", command = runBar)
    btn.pack(pady = 10)    

#--------------------------------------------------------------------------------------------------------------------------------------------
'''SEARCH WINDOW'''

def openSearchBook():
      
    # Toplevel object which will 
    # be treated as a new window
    newWindow = Toplevel(master)
  
    # sets the title of the
    # Toplevel widget
    newWindow.title("Search Book")
  
    # sets the geometry of toplevel
    newWindow.geometry("600x300")
  

    # A Label widget to show in toplevel
    Label(newWindow, 
          text ="This is a new window").pack()

#--------------------------------------------------------------------------------------------------------------------------------------------
'''MAIN WINDOW'''

label = Label(master, 
              text ="Welcome to the MSP Database")
  
label.pack(pady = 10)
  
# a button widget which will open a 
# new window on button click
btn = Button(master, 
             text ="Add Book", 
             command = openAddBook)
btn.pack(pady = 10)

btn2 = Button(master, 
             text ="Search Book", 
             command = openSearchBook)
btn2.pack(pady = 10)
  
# mainloop, runs infinitely
mainloop()
```
